Test using Amadeus API. We need to crete an account and get the credentials.

```python
from amadeus import Client, ResponseError
import os
from dotenv import load_dotenv

load_dotenv()



# Replace with your actual API Key and API Secret
AMADEUS_API_KEY = os.getenv("AMADEUS_API_KEY")
AMADEUS_API_SECRET = os.getenv("AMADEUS_API_SECRET")

amadeus_client = Client(
    client_id=AMADEUS_API_KEY,
    client_secret=AMADEUS_API_SECRET
)

try:
    response = amadeus_client.shopping.flight_destinations.get(
        origin='PAR',
        maxPrice=200,
        currency='EUR'
    )
    print(response.data)
    # => [{'type': 'checkin-link', 'id': '1XEN-GBWeb', 'href': 'https://www....
except ResponseError as error:
    print(error)
```

```python
print(type(response))
print(type(response.data))
print(len(response.data))
print(response.data[0])
print(type(response.data[0]))
print(response.data[0]['type'])
print(response.result)
```

```python
def get_flights_from_to(origin, destination, departure_date, number_of_adults):
    try:
        response = amadeus_client.get('/v2/shopping/flight-offers',
            originLocationCode=origin,
            destinationLocationCode=destination,
            departureDate=departure_date,
            adults=number_of_adults
        )
        return response
    except ResponseError as error:
        print(f"Error fetching flight offers: {error}")
        raise error
```

```python
r = get_flights_from_to(origin='MUC', destination='UIO', departure_date='2025-05-28', number_of_adults=2)
```
